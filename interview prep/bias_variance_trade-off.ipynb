Error(Model) = Variance(Model) + Bias(Model) + Variance(Irreducible Error)

### Bias
- Low Bias: Weak assumptions regarding the functional form of the mapping of inputs to outputs.
- High Bias: Strong assumptions regarding the functional form of the mapping of inputs to outputs.

### Variance

- Low Variance: Small changes to the model with changes to the training dataset.
- High Variance: Large changes to the model with changes to the training dataset.

### Irreducible Error

- Model Error = Reducible Error + Irreducible Error

### Boston Housing dataset example

#### Data prep

```python
from mlxtend.evaluate import bias_variance_decomp
from mlxtend.data import boston_housing_data
from sklearn.model_selection import train_test_split

X, y = boston_housing_data()
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=123,
                                                    shuffle=True)
```

#### Fitting models

##### Linear Regression

```python
from sklearn.linear_model import LinearRegression

# define the model
linearReg = LinearRegression()

# estimate bias and variance
avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        linearReg, X_train, y_train, X_test, y_test, 
        loss='mse',
        num_rounds=200,
        random_seed=123)

# summarize results
print('Average expected MSE: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

```python
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor(random_state=123)

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        tree, X_train, y_train, X_test, y_test, 
        loss='mse',
        num_rounds=200,
        random_seed=123)

print('Average expected MSE: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

####  Bias Variance Decomposition of eXtreme Gradient Boosting regressor

```python
import xgboost as xgb

from sklearn.model_selection import KFold, GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_boston

boston = load_boston()
y = boston['target']
X = boston['data']
kf = KFold(n_splits=2, shuffle=True, random_state=123)
for train_index, test_index in kf.split(X):
    xgb_model = xgb.XGBRegressor().fit(X[train_index], y[train_index])
    predictions = xgb_model.predict(X[test_index])
    actuals = y[test_index]
    print(mean_squared_error(actuals, predictions))
```

### Bias Variance Decomposition of a Voting Classifier

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier

clf1 = LogisticRegression(multi_class='multinomial', random_state=123)
clf2 = RandomForestClassifier(random_state=123)

tree = DecisionTreeClassifier(random_state=123)
clf3 = BaggingClassifier(base_estimator=tree,
                        n_estimators=100,
                        random_state=123)
clf4 = KNeighborsClassifier(n_neighbors=10)

eclf1 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('bag', clf3), ('knn', clf4)], voting='hard')

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        eclf1, X_train, y_train, X_test, y_test, 
        loss='0-1_loss',
        random_seed=123)


print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

```python
eclf2 = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('bag', clf3), ('knn', clf4)], voting='soft')

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        eclf2, X_train, y_train, X_test, y_test, 
        loss='0-1_loss',
        random_seed=123)


print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
